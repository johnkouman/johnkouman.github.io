```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('emdat_public_2023_03_10_query_uid-vgqrHL.csv', header=6,encoding='cp1252')
```

```python
df
```

```python
count_year = df.groupby(['Year','Disaster Type'])['Disaster Type'].count().reset_index(name='Count_year')
```

```python
count_year
```

```python
count_year['Count_year'].sum()
```

```python
plt.figure(figsize=(15, 30))
count_year.plot.bar(y='Count_year', x='Year')
plt.xticks(rotation=90);
```
